```python
import pandas as pd
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data.csv")
df.rename(columns=lambda x: x[0].lower() + x.strip().lower().replace(' ', '_')[1:], inplace=True)
df = df.convert_dtypes()
df
```

Meet Eleanor. She’s an American college grad who values cultural diversity. She wants to go to grad school, but she’s having trouble sorting through all the college reports to figure out which one(s) she will apply to. So much information! It’s overwhelming!

This is a big problem. She can only afford application fees for 3 schools. This is making her very nervous.

So, she’s going to play the numbers game in order to quantify all this information and make an educated decision. She’s just that kind of person -- a science nerd.

First, she knows she’s gotta figure out what she wants. She values cultural diversity. As a prospective environmental science major, sustainability is a must. She also wants to make sure that her tuition brings her the best bang for her buck, so professionalization and networking are really important to her.

```python
sns.relplot(df)
```

A list of over 1500 schools, to which she plans on only applying to a handful, is simply too much data to root through by hand.

She decides to identify her values first.

*   Cultural diversity
*   Environmental consciousness
*   Professional development






```python
from matplotlib import pyplot as plt
df.plot(kind='scatter', x='international_faculty', y='international_students', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

There is no apparent correlation between international faculty and international student populations.

```python
sns.catplot(x='location', y="institution_name", data=df, kind='bar')
```

Eleanor decides that schools in France would be her best option, based on the criteria she has laid out. This makes a lot of sense!





```python
#New dataset for all schools in France
France = df[df.location=="FR"]
France
```

```python
# @title Sustainability in French schools

from matplotlib import pyplot as plt
France['sustainability'].plot(kind='hist', bins=20, title='sustainability')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
from matplotlib import pyplot as plt
France['employment_outcomes'].plot(kind='hist', bins=20, title='employment outcomes')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

Now that Eleanor has all the data she needs organized and displayed, it's time to put it all together!

Since there are 4 schools in the top 100 that match her criteria, Eleanor decides to choose 4 instead of just to which to apply.

```python
# top schools in France

sns.catplot(x="institution_name", y="2025_rank", data=France, kind="bar")
plt.xlim(0, 3)
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.show()
```

Success!

Eleanor has identified the top schools to which to apply.

1.   Sorbonne University
1.   Université PSL
2.   Université Paris-Saclay
1.   Institut Polytechnique de Paris

And she lived happily ever after.
