```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("data.csv")
df.rename(columns=lambda x: x[0].lower() + x.strip().lower().replace(' ', '_')[1:], inplace=True)
df = df.convert_dtypes()
df
```

```python
df["American"] = df.location == "US"
df
```

```python
IsForeignSchBenefit=sns.boxplot(data=df, x="American",y="employment_outcomes")
IsForeignSchBenefit.set_xlabel("Country of School")
IsForeignSchBenefit.set_xticklabels(["Non-US", "US"])
```

This unfiltered plot shows that there is no clear job-finding advantage to a country change for an American student, but there are a lot of outliers in both groups of schools.

```python
SizeImpactSust = sns.catplot(data=df, x = "size", y = "sustainability", order= ["S", "M", "L", "XL"], hue="size")
```

This plot shows that, in fact, there is a clear positive relationship between the size of a school and its sustainability efforts, at least for small and medium size schools. There seems to be less of an impact going from large to extra large. You might be able to assume that is because once they reach a certain size, they have the money to spend on sustainability.
